```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)
```

```python
means = [[0, 0], [0, 0]]
cov0 = [[1.02, 1], [1, 1.02]]
cov1 = [[1.03, 1], [1, 1.03]]
N0, N1 = 40, 50
X0 = np.random.multivariate_normal(means[0], cov0, N0)
X1 = np.random.multivariate_normal(means[1], cov1, N1)
X = np.concatenate((X0, X1))
y = np.array([0]*N0 + [1]*N1)
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label = 'class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label = 'class 1')

plt.legend()
plt.show()
```

```python
X_mean = np.mean(X.T, axis = 1, keepdims = True)
C0 = X0.T - np.tile(X_mean[0], (1, N0))
C1 = X1.T - np.tile(X_mean[1], (1, N1))
S = (C0.dot(C0.T) + C1.dot(C1.T))/(N0 + N1)
val, vec = np.linalg.eig(S)
print(val)
print(vec)
print(np.tile(X_mean[0], (1, N0)))
```

```python
Y = lambda x1: -x1*vec[0][0]/vec[0][1]
Y_ = lambda x1: -x1*vec[1][0]/vec[1][1]
X2_ = Y(X0[:, 0])
X2__ = Y_(X0[:, 0])

plt.figure(figsize=(5, 5))
plt.scatter(X[y == 0][:, 0], X[y == 0][:, 1], label = 'class 0')
plt.scatter(X[y == 1][:, 0], X[y == 1][:, 1], label = 'class 1')
plt.plot(X0[:, 0], X2_, color = 'red', label = 'vec 0')
plt.plot(X0[:, 0], X2__, color = 'green', label = 'vec 1')
plt.legend()
```
